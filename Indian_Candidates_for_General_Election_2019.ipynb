```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
%matplotlib inline
```

```python
df = pd.read_csv('LS_2.0.csv')
df.head()
```

```python
df.columns
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,5))
chart= sns.countplot(df["STATE"])
l = list(df['STATE'].unique())
chart.set_xticklabels(labels=l, rotation=90)
```

```python
sns.countplot(df['CATEGORY'])
```

```python
plt.figure(figsize=(40,10))
chart= sns.countplot(df["PARTY"])
l = list(df['PARTY'].unique())
chart.set_xticklabels(labels=l, rotation=90)
```

```python
plt.figure(figsize=(90,50))
pie = df['STATE'].value_counts().plot(kind="pie", autopct='%1.1f%%',
        shadow=True, startangle=90 )
plt.legend(loc='right')
pie.set_ylabel('')
```

```python
sns.set(font_scale=1)
barplot1 = df['GENDER'].value_counts(ascending=False).plot(kind='bar')
plt.xlabel("Gender", labelpad=10)
plt.ylabel("Count", labelpad=10)
plt.show()
```
