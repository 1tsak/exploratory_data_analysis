```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
df=  pd.read_csv("./data/Data_Science_Jobs_in_India.csv", index_col=0)
df.head()
```

```python
df.tail()
```

```python
#making a Dataframe of top 5 service based and product based companies
df5_both = df[(df.company_name.isin(['TCS','Infosys','Wipro','Accenture','Cognizant'])) | (df.company_name.isin(['Dell','Amazon','Apple','PayPal','Google']))]
df5_service = df[df.company_name.isin(['TCS','Infosys','Wipro','Accenture','Cognizant'])]
df5_product = df[df.company_name.isin(['Dell','Amazon','Apple','PayPal','Google'])]
```

```python
df5_both.avg_salary = df5_both.avg_salary.str.replace('L', '').values.astype(np.float64)
df5_both.min_salary = df5_both.min_salary.str.replace('L', '').values.astype(np.float64)
df5_both.max_salary = df5_both.max_salary.str.replace('L', '').values.astype(np.float64)


df5_service.avg_salary = df5_service.avg_salary.str.replace('L', '').values.astype(np.float64)
df5_service.min_salary = df5_service.min_salary.str.replace('L', '').values.astype(np.float64)
df5_service.max_salary = df5_service.max_salary.str.replace('L', '').values.astype(np.float64)


df5_product.avg_salary = df5_product.avg_salary.str.replace('L', '').values.astype(np.float64)
df5_product.min_salary = df5_product.min_salary.str.replace('L', '').values.astype(np.float64)
df5_product.max_salary = df5_product.max_salary.str.replace('L', '').values.astype(np.float64)
```

```python
df5_both.reset_index(drop=True)
df5_service.reset_index(drop=True)
```

```python
df5_product.reset_index(drop=True)
```

```python
df5_service.describe()
```

```python
df5_product.describe()
```

```python
df5_service.job_title.unique()
```

```python
df5_product.job_title.unique()
```

```python
# Based on above two cells, it is evident that both the types have same roles available except Data Architect which is
# only present in Service Based compnaies
```

```python
df5_service.groupby('job_title').min_experience.mean().round(2)
```

```python
df5_product.groupby('job_title').min_experience.mean().round(2)
```

```python
# Based on above two cells, The average number of years of experience for Machine Learning Engineer, Senior Business
# Analysist and Senior Data Analyst is less in product compnaies than service companies.
# Except these three, it requires more or same number of experience for a role in product companies than in service companies
```

```python
df5_service.groupby('job_title').avg_salary.mean().round(2)
```

```python
df5_product.groupby('job_title').avg_salary.mean().round(2)
```

```python
#  Quite clear that product companies pay higher than service based companies for each role
```

```python
df5_service.groupby('job_title').num_of_salaries.mean().round(2)
```

```python
df5_product.groupby('job_title').num_of_salaries.mean().round(2)
```

```python
# As per the dataset, the number of people working at same role is higher in service companies than in product companies
```
