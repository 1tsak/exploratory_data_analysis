```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
data = pd.read_csv('data/crime_by_state.csv')
data.head()
```

**DATA CLEANING**

```python
gen_order = data['STATE/UT'].value_counts().index
print(gen_order)
data.isnull().sum()
```

**DATA VISUALISATION**

```python
WB_data = data[data["STATE/UT"] == "WEST BENGAL"]
WB_data.isnull().sum()
WB_data["Year"]
```

```python
WB_data["Total Crimes"]  = WB_data.sum(axis=1, numeric_only=True)
WB_data["Year"] = WB_data["Year"].astype(int)
WB_data["Total Crimes"] = WB_data["Total Crimes"].astype(int)
WB_data["Total Crimes"] = WB_data['Total Crimes'] - WB_data["Year"]
WB_data
```

```python
state_df = WB_data
result = state_df.groupby("Year")["Total Crimes"].sum()
Yearly_Crimes = {}
for year, records in result.items():
    Yearly_Crimes[year] = records
Yearly_Crimes = pd.DataFrame({'Year' : Yearly_Crimes.keys() , 'Total Crimes' : Yearly_Crimes.values()})
base_color = sb.color_palette()[0]
sb.barplot(x = "Year",y = "Total Crimes",data = Yearly_Crimes,color = base_color)
plt.title("Increase in Crimes over the years in {state}".format(state = "West Bengal"))
```
